```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
df = pd.read_csv("Dynamic_weight.csv")
print(df.columns)
epsilon = list(df['epsilon'])
rt_s1 =list(df['runtime(S1)']) 
rt_m1 =list(df['runtime(M1)'])
rt_l1 =list(df['runtime(L1)'])
en_s1 =list(df['Nodes(S1)'])  
en_m1 =list(df['Nodes(M1)'])
en_l1 =list(df['Nodes(L1)'])
pm_s1 =list(df['PM2.5(S1)'])
pm_m1 =list(df['PM2.5(M1)'])
pm_l1 =list(df['PM2.5(L1)'])
```

# L1

```python
f1, axs = plt.subplots(1,3,figsize=(9,3))
axs[0].plot(epsilon, pm_l1,label = 'dynamic-weight A*')
axs[0].set_xlabel('epsilon')
axs[0].set_ylabel('pm 2.5 ' + r'$(\mu g/m^3)$')
axs[0].plot(0, 67.53, marker='o', markersize=10, color="red", label = 'A*')
axs[0].legend(loc="lower right", prop={'size': 8})

axs[1].plot(epsilon, en_l1, label = 'dynamic-weight A*')
axs[1].set_xlabel('epsilon')
axs[1].set_ylabel('expanded nodes')
axs[1].plot(0, 63339, marker='o', markersize=10, color="red", label = 'A*')
axs[1].legend(loc="upper right", prop={'size': 8})

axs[2].plot(epsilon, rt_l1, label = 'dynamic-weight A*')
axs[2].set_xlabel('epsilon')
axs[2].set_ylabel('run time (ms)')
axs[2].plot(0, 67, marker='o', markersize=10, color="red", label = 'A*')
axs[2].legend(loc="upper right", prop={'size': 8})

f1.tight_layout()
plt.savefig("DW_L1.pdf")
```
