```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import matplotlib
```

```python
df = pd.read_csv("lpr_short_day_2019_07_24.csv")
print(df.columns)
```

```python
time = list(df['Time'])
PM_lpr = list(df['PM2.5(lpr)'])
PM_short = list(df['PM2.5(short)'])
Distance_lpr = list(df['Distance(lpr)'])
Distance_short = list(df['Distance(short)'])
print(PM_lpr)
print(PM_short)
```

```python

xi = [i for i in range(0, len(time))]
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10, 5)
plt.plot(PM_lpr, label="least polluted route")
plt.plot(PM_short, label="shortest route")
plt.xticks(xi, xi)
plt.xlabel("Time")
plt.ylabel("pm 2.5" + r'$(\mu g/m^3)/(mile)$')
plt.legend(loc="upper right")
plt.savefig("lpr_short_PM.pdf")
```

```python
xi = [i for i in range(0, len(time))]
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10, 5)
plt.plot(Distance_lpr,marker='o', linestyle='dashed', linewidth=1, markersize=8, label="least polluted route")
plt.plot(Distance_short,marker='o', linestyle='dashed', linewidth=1, markersize=8, label="shortest route")
plt.xticks(xi, xi)
plt.xlabel("Time")
plt.ylabel("Distance(mile)")
plt.legend(loc="upper right")
plt.savefig("lpr_short_distance.pdf")
```
